```python
# Create annualized dataset specifically for Iraq
gdp = gdp_info_raw[gdp_info_raw["Country Name"] == "Iraq"].transpose()[4:-1]
gdp.columns = ["GDP"]
gdp.index = pd.to_numeric(gdp.index)
gdp.index.name = "Year"
gdp["GDP"] = pd.to_numeric(gdp["GDP"])
gdp.info()
```

```python
brd = pd.read_csv("data/UDCP/ucdp-brd-dyadic-2014.csv")
brdi = brd[brd["LocationInc"] == "Iraq"]
brdi = brdi.groupby("Year").sum().drop(["ConflictID","DyadId","SideBID","Incompatibility","TypeOfConflict"],1)
brdi.info()
```

```python
hcrpsq = pd.read_csv("data/unhcr_popstats_export_time_series_all_data.csv",skiprows=range(2),na_values="*")
#hcrpsq = hcrpsq[hcrpsq["Population type"] != "Returnees"][hcrpsq["Population type"] != "Returned IDPs"]
hcrpsq = hcrpsq[hcrpsq["Population type"] != "Returnees"]
hcrpsq = hcrpsq[hcrpsq["Population type"] != "Returned IDPs"]

fhcrpsq = hcrpsq[hcrpsq["Origin"] == "Iraq"]

rdata = fhcrpsq[fhcrpsq["Population type"] == "Refugees (incl. refugee-like situations)"].groupby('Year').sum() #rdata = fhcrpsq[fhcrpsq["Population type"] != "Persons in Refugee-like situation"].groupby('Year').sum()
rdata.columns = ["Refugees + IDP"]
rdata["Previous Year Refugees"] = rdata["Refugees + IDP"].shift(1)
rdata["Delta Refugees"] = rdata["Refugees + IDP"].shift(2) - rdata["Refugees + IDP"].shift(1)
rdata["4 yr avg Refugees"] = (rdata["Refugees + IDP"].shift(4) + rdata["Refugees + IDP"].shift(3) + rdata["Refugees + IDP"].shift(2) + rdata["Refugees + IDP"].shift(1))/4
rdata.index.name = "Year"
rdata.info()
print hcrpsq["Population type"].unique()

#fhcrpsq
```

```python
unemp = pd.read_csv("data/world_bank_unemployment/API_SL.UEM.TOTL.ZS_DS2_en_csv_v2.csv",
                     skiprows=range(3), 
                     na_values="*"
                   ).set_index("Country Name").transpose()[3:-1][["Iraq"]]
unemp.columns = ["Unemployment"]
unemp.index = unemp.index.astype(np.int64)
unemp.info()
```

```python
cpd = pd.read_table("data/centripetalism.txt")
cpdi = cpd[cpd["Country"] == "Iraq"]
cpdid = cpdi[["Year",
              "GDPpc_imputed_ln_wtradeF",
              "Growthpc_imputed_sd5_wgeoF",
              "IMR_ipolate_ln_wgeoF",
              "Oil_production_MH_pc",
              "Participation_Vanhanen_wgeoF",
              "Polity4_imp_sd5_wgeoF",
              "Schooling_tot_ipo_wgeoF",
              "Stock_depr1",
              "Tax_Revenue_WDI_wgeoF",
              "Trade_WDI_ln_wgeoF"]]
cpdid = cpdid.groupby("Year").sum()
cpdid.info()
```

```python
concat = rdata.join([gdp,brdi,unemp,cpdid], how="outer")
concat["BdBest"] = concat["BdBest"].fillna(0)
concat["BdLow"] = concat["BdLow"].fillna(0)
concat["BdHigh"] = concat["BdHigh"].fillna(0)
concat
```

```python
from sklearn.cross_validation import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import Imputer
from sklearn.svm import LinearSVR
from sklearn import ensemble
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
```

```python
cpruned = concat[np.isfinite(concat["Refugees + IDP"])][13:]

features = cpruned[["Previous Year Refugees",
                    "Delta Refugees", 
                    "4 yr avg Refugees", 
                    "GDP",
                    "BdBest",
                  "Participation_Vanhanen_wgeoF",
                  "Schooling_tot_ipo_wgeoF",
                  "Tax_Revenue_WDI_wgeoF",
                  "Trade_WDI_ln_wgeoF"]]
target = cpruned["Refugees + IDP"]

cpruned
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
features_filled = imp.fit_transform(features)


features_train, features_test, target_train, target_test = train_test_split(features_filled, target, test_size=0.2, random_state=198902423)

min_max_scaler = preprocessing.MinMaxScaler()

features_train_mm = min_max_scaler.fit_transform(features_train)
features_test_mm = min_max_scaler.transform(features_test)

clf = ensemble.AdaBoostRegressor(random_state=0, n_estimators=80)

clf.fit(features_train_mm,target_train)

prediction = clf.predict(features_test_mm)

rm = np.sqrt(metrics.mean_squared_error(target_test,prediction))

print rm

print 1 - rm / np.mean(target_test)
pd.DataFrame(zip(features_test, target_test, prediction, (target_test - prediction)/target_test))
```
